```python
#sentiment analysis(subjectivity and polarity)
```

```python
import pandas as pd 
from textblob import TextBlob
```

```python
data=pd.read_csv(r"C:\computer\amazon_alexa.tsv", delimiter="\t")
```

```python
data.shape
```

```python
data.head(5)
```

```python
data['length']=data['verified_reviews'].apply(len)
```

```python
data.head()
```

```python
# text polarity says sentiment is positive negative or neutral speaking good or bad in reviews
```

```python
#lets calculate polarity of the reviews
def get_polarity(text):
    textblob=TextBlob(str(text.encode('utf-8')))
    pol=textblob.sentiment.polarity
    return pol
# let's apply it to a function
data['polarity']=data['verified_reviews'].apply(get_polarity)
```

```python
data.head()
```

### subjectivity 

subjectivity refers how strong are your opinion how much positive or negative

```python
# let's calculate subjectivity of reviews
def get_subjectivity(text):
    textblob=TextBlob(str(text.encode('utf-8')))
    subj=textblob.sentiment.subjectivity
    return subj
#let's apply function by creating new column
data['subjectivity']=data['verified_reviews'].apply(get_subjectivity)
```

```python
data.head()
```

```python
data[["length","polarity","subjectivity"]].describe()
```

# finding words, character and punctuation marks

```python
data=pd.read_csv(r"C:\computer\amazon_alexa.tsv", delimiter="\t")
```

```python
data['char_count']=data['verified_reviews'].apply(len)
```

```python
data['word_count']=data['verified_reviews'].apply(lambda x:len(x.split()))
```

```python
#calculate word density
data['word_density']=data['char_count']/(data['word_count']+1)
```

```python
import string
```

```python
punctuation
```

```python
data['punctuation_count']=data['verified_reviews'].apply(lambda x:len("".join(_ for _ in x if _ in punctuation)))
```

```python
from textblob import TextBlob
text= ("codespeedy is aprogramming blog"
      "blog post contain articles tutorials and programs on css html and javascript")
tb=TextBlob(text)
print(tb.tags)
```

```python
# lets create parts of speech dictionary
pos_dic={
    'noun':['NN','NNS','NNP','NNPS'],
    'pronoun':['PRP','PRP$','WP','WP$'],
    'verb':['VB','VBD','VBG','VBN','VBN','VBZ'],
    'adjective':['JJ','JJR','JJS'],
    'adverb':['RB','RBR','RBS','WRB']
    
}
#function to check and get the part of speech tag count of words in a given sentence
def pos_check(x,flag):
    cnt=0
    try:
        wiki=TextBlob(x)
        for tup in wiki.tags:
            ppo=list(tup)[1]
            if ppo in pos_dic[flag]:
                cnt+=1
    except:
        pass
    return cnt
```

```python
pos_check(text,"noun")
```

```python
data['noun_count']=data['verified_reviews'].apply(lambda x: pos_check(x, "noun"))
```

```python
data['verb_count']=data['verified_reviews'].apply(lambda x: pos_check(x, "verb"))
```

```python
data[['noun_count','verb_count']].describe()
```
